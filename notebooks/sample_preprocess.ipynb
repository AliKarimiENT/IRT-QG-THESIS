# Sample Data Preprocessing Notebook

This notebook demonstrates how to use the data preparation utilities for the thesis project.

## Overview
- Load raw data
- Clean and preprocess data
- Save processed data
- Generate visualizations


```python
# Import necessary libraries
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Add src directory to path to import project modules
sys.path.append(str(Path.cwd().parent / "src"))

# Import project modules
from config import RAW_DATA_DIR, PROCESSED_DATA_DIR, FIGURES_DIR
from data.prepare import load_raw_data, clean_data, preprocess_data, save_processed_data
from utils.io_utils import save_dataframe, load_dataframe

# Set up plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("Libraries imported successfully!")
print(f"Raw data directory: {RAW_DATA_DIR}")
print(f"Processed data directory: {PROCESSED_DATA_DIR}")
print(f"Figures directory: {FIGURES_DIR}")
```

## 1. Load Raw Data

First, let's check what data files are available in the raw data directory.

```python
# Check what files are in the raw data directory
raw_files = list(RAW_DATA_DIR.glob("*"))
print("Files in raw data directory:")
for file in raw_files:
    print(f"  - {file.name}")

# If no files exist, create a sample dataset for demonstration
if not raw_files:
    print("\nNo files found. Creating a sample dataset for demonstration...")
    
    # Create sample data
    np.random.seed(42)
    n_samples = 1000
    
    sample_data = pd.DataFrame({
        'id': range(1, n_samples + 1),
        'feature_1': np.random.normal(0, 1, n_samples),
        'feature_2': np.random.normal(5, 2, n_samples),
        'feature_3': np.random.choice(['A', 'B', 'C'], n_samples),
        'target': np.random.choice([0, 1], n_samples, p=[0.7, 0.3])
    })
    
    # Add some missing values for demonstration
    sample_data.loc[sample_data.sample(50).index, 'feature_1'] = np.nan
    sample_data.loc[sample_data.sample(30).index, 'feature_2'] = np.nan
    
    # Save sample data
    sample_data.to_csv(RAW_DATA_DIR / "sample_data.csv", index=False)
    print(f"Created sample dataset with {n_samples} rows and saved to sample_data.csv")
    
    # Display basic info about the sample data
    print("\nSample data info:")
    print(sample_data.info())
    print("\nFirst few rows:")
    print(sample_data.head())
```

## 2. Data Cleaning and Preprocessing

Now let's use our data preparation functions to clean and preprocess the data.


```python
# Load the raw data
raw_data = load_raw_data("sample_data.csv")
print(f"Raw data shape: {raw_data.shape}")
print(f"Missing values:\n{raw_data.isnull().sum()}")

# Clean the data
cleaned_data = clean_data(raw_data)
print(f"\nCleaned data shape: {cleaned_data.shape}")

# Preprocess the data
processed_data = preprocess_data(cleaned_data)
print(f"Processed data shape: {processed_data.shape}")

# Save the processed data
save_processed_data(processed_data, "processed_sample_data.csv")
print("Processed data saved successfully!")
```

## 3. Data Visualization

Let's create some visualizations to explore the data.


```python
# Create visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Distribution of feature_1
axes[0, 0].hist(processed_data['feature_1'].dropna(), bins=30, alpha=0.7, color='skyblue')
axes[0, 0].set_title('Distribution of Feature 1')
axes[0, 0].set_xlabel('Feature 1')
axes[0, 0].set_ylabel('Frequency')

# Distribution of feature_2
axes[0, 1].hist(processed_data['feature_2'].dropna(), bins=30, alpha=0.7, color='lightcoral')
axes[0, 1].set_title('Distribution of Feature 2')
axes[0, 1].set_xlabel('Feature 2')
axes[0, 1].set_ylabel('Frequency')

# Categorical distribution
feature_3_counts = processed_data['feature_3'].value_counts()
axes[1, 0].bar(feature_3_counts.index, feature_3_counts.values, color='lightgreen')
axes[1, 0].set_title('Distribution of Feature 3 (Categorical)')
axes[1, 0].set_xlabel('Feature 3')
axes[1, 0].set_ylabel('Count')

# Target distribution
target_counts = processed_data['target'].value_counts()
axes[1, 1].bar(target_counts.index, target_counts.values, color='gold')
axes[1, 1].set_title('Target Distribution')
axes[1, 1].set_xlabel('Target')
axes[1, 1].set_ylabel('Count')

plt.tight_layout()
plt.savefig(FIGURES_DIR / 'data_exploration.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"Visualization saved to {FIGURES_DIR / 'data_exploration.png'}")
```
